## Goal

## API setup

````python
from src.kalshi_client import KalshiClient

client = KalshiClient()
````

## Search GDP events

````python
events = client.search_gdp_events()
events[["event_ticker", "title", "sub_title", "series_ticker"]].head(20)
````

## Identify Q2 2026 event

````python
event = client.find_event_by_title("US GDP growth in Q2 2026?", series_ticker="KXGDP")
event
````

## Pull market contracts

````python
markets = client.list_markets(event_ticker=event["event_ticker"])
thresholds = client.add_threshold_probabilities(markets)
thresholds[["ticker", "yes_sub_title", "threshold", "yes_mid", "last_price_prob", "prob_above"]]
````
